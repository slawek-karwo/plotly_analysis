```python
# import goal scorrers of PGNiG Superliga from api: http://rozgrywki.zprp.pl/api/ (polish official handball website)
```

```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
```

```python
# actual season is 9857 so I use that api request: http://rozgrywki.zprp.pl/api/pokaz_strzelcow.php?Rozgrywki=9857
# we can see that also on webpage: https://rozgrywki.zprp.pl/?Sezon=188&Rozgrywki=9857&Tabela3=1
url = r'http://rozgrywki.zprp.pl/api/pokaz_strzelcow.php?Rozgrywki=9857'
r = requests.get(url)
# print(r.content)
print(r.text)
```

```python
scorrers = json.loads(r.text)
print(scorrers)
```

```python
# parameter ensure_ascii=False
# If ensure_ascii is true (the default), the output is guaranteed to have all incoming non-ASCII characters escaped. 
# If ensure_ascii is false, these characters will be output as-is.
# delete firts row (useless)
scorrers.pop(0)
json_formatted = json.dumps(scorrers, indent=1, ensure_ascii=False)
print(json_formatted)
```

```python
scorrers_table = pd.read_json(json_formatted)
scorrers_table
```

```python
scorrers_table[scorrers_table['karne_liczba'] == scorrers_table['karne_liczba'].max()]
```

```python
scorrers_table.sort_values(by='karne_liczba', ascending=False)
```
